```python
def obter_opcao():
    opcao = 0
    while opcao not in range(1, 9):
        try:
            opcao = int(input("Primeiro, escolha uma opção digitando um número de 1 a 8: "))
            if opcao in range(1, 9):
                return opcao
            else:
                print("Opção inválida. Escolha um número entre 1 e 8.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número inteiro entre 1 e 8")

```

```python
def obter_numeros():
    while True:
        try:
            x = float(input("Digite o primeiro número: "))
            y = float(input("Digite o segundo número: "))
            return x, y
        except ValueError:
            print("Entrada inválida. Por favor, insira números válidos.")
```

```python
def realizar_calculo(opcao, x, y):

    if opcao == 1:
        print(f"A soma de {x} e {y} é {x + y}")
    elif opcao == 2:
        print(f"A subtração de {x} e {y} é {x - y}")
    elif opcao == 3:
        print(f"A multiplicação de {x} e {y} é {x * y}")
    elif opcao == 4:
        if y != 0:
            print(f"A divisão de {x} por {y} é {x / y}")
        else:
            print("Divisão por zero não é permitida.")
    elif opcao == 5:
        if y != 0:
            print(f"A divisão inteira de {x} por {y} é {x // y}")
        else:
            print("Divisão por zero não é permitida.")
    elif opcao == 6:
        if y != 0:
            print(f"O resto da divisão de {x} por {y} é {x % y}")
        else:
            print("Divisão por zero não é permitida.")
    elif opcao == 7:
        print(f"A potência de {x} elevada a {y} é {x ** y}")
    elif opcao == 8:
        print(f"A soma de {x} e {y} é {x + y}")
        print(f"A subtração de {x} e {y} é {x - y}")
        print(f"A multiplicação de {x} e {y} é {x * y}")
        if y != 0:
            print(f"A divisão de {x} por {y} é {x / y}")
            print(f"A divisão inteira de {x} por {y} é {x // y}")
            print(f"O resto da divisão de {x} por {y} é {x % y}")
        else:
            print("Divisão por zero não é permitida.")
        print(f"A potência de {x} elevada a {y} é {x ** y}")
```
